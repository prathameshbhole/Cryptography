```python
#Extended euclidean algorithm: ax + by = gcd(a,b)
def ExtendedEuclidean(a,b):
    #Base case
    if(a == 0):
        return b, 0, 1
    else:
        gcd, x, y = ExtendedEuclidean(b%a, a)
        return (gcd, y - (b//a)*x, x)

#a function that finds the multiplicative inverse of a mod n
def MultiplicativeInverse(a,n):
    gcd, x, y = ExtendedEuclidean(a,n)
    #A multiplicative inverse only exists if a and n are relatively prime, so gcd must be 1
    if(gcd == 1):
        #Get rid of negative values
        if(x < 1): 
            return x % n
        return x
    return 0

#gcd
def gcd(a, b):
    if a == 0 :
        return b
    return gcd(b%a, a)

#Two random prime numbers
p=1
q=1
def prime():
    global p,q
    print('\nTwo random prime numbers:')
    p=int(input(print('\np=')))
    q=int(input(print('\nq=')))
    if p==q:
        print('Please enter different numbers')
        prime()
    
prime()
for i in range(2,p):
        if p%i==0:
            print('\nEnter the numbers again')
            prime()
for i in range(2,q):
        if q%i==0:
            print('\nEnter the numbers again')
            prime()
#product of prime numbers
n=p*q
phin = (p-1)*(q-1)
print("\nn = ",n,"\nphi(n) = ",phin)

#selection of e
e=2
j=0
for i in range(2,phin):
    global e,p,q
    if (gcd(e,phin)==1 and e!=p and e!=q):
        global j
        j=j+1
        if j==2:
            break
        else:e+=1
    else: e+=1

#selection of d
d= MultiplicativeInverse(e,phin)
print('e =',e,'d =',d)

M=int(input(print('Enter the Message')))

#message
def msg(): 
    global M 
    M=int(input(print('\nEnter the Message again less than<',n)))

if (M>n or M==n): msg()

#cipher text
C=int(pow(M,e,n))
print('\nCipher text:',C)

#decryption
Msg=int(pow(C,d,n))
print('\nDecrypted message:',Msg)
```
